### Temperature fluctuations are directly related to density perturbations of baryon plasma

Because before recombination thermal radiation and baryons are coupled by Thompson scattering, their fluctuations are related.

$$\frac{\Delta T}{\bar{T}} \approx \frac{1}{3}\,\frac{\Delta\rho_{\rm bar}}{\bar{\rho}_{\rm bar}}$$

where $\Delta\rho_{\rm bar}/\bar{\rho}_{\rm bar}$ is fluctuation amplitude of the *baryon* matter (i.e., normal nuclei, atoms, and electrons) and $\bar{\rho}_{\rm bar}$ is the mean density of baryon matter in the universe. 

*Thus, fluctuations of CMB temperature reveal perturbations in baryon density* at the last epoch when CMB radiation and baryon plasma were coupled (redshift of $z_{\rm ls}\approx 1100$)

### "Lagrangian" scale of a perturbation of mass $M$

We can define *comoving Lagrangian radius* corresponding to mass scale $M$ as

$$R_{\rm L}=\,\left[\frac{3M}{4\pi\bar{\rho}_{\rm m0}}\right]^{1/3}\approx 1.8\ {\rm Mpc}\,\left(\frac{M}{10^{12}\, M_\odot}\right)^{1/3}\,\left(\frac{\Omega_{\rm m0}}{0.3}\right)^{-1/3}\,\left(\frac{h}{0.7}\right)^{-2/3}$$


The observed collapsed structures (clusters, groups, galaxies) with masses $M\leq 10^{15}\ M_\odot$ thus have $R_L\leq 20$ Mpc. 

A physical scale $R_L/(1+z_{\rm ls})$ is related to observed angular size $\theta_L$ via the *angular diameter distance*, $d_{\rm A}$:

$$\theta_L\approx \frac{D_L}{d_{\rm A}(z_{\rm ls}, \Omega_{\rm m0}, \Omega_{\Lambda 0})}\, \frac{180^{\circ}}{\pi}\approx 0.1^{\circ}\,\frac{R_{L}}{20\,\rm Mpc},$$

Thus, to get a sense of the initial amplitude of density fluctuations, we should thus look at scales $\theta\leq 0.1^{\circ}$.

### The implied amplitude of baryon fluctuations

On scales $\theta\leq 0.1-0.2^\circ$, $\Delta_{\rm T}^2\approx 200\ (\mu K)^2$. So typical temperature fluctuation is

$$\frac{\Delta T}{\langle T\rangle}\approx \frac{\sqrt{200\times 10^{-12}\, K^2}}{2.678\, K}\approx 5\times 10^{-6}$$. 

```python
print(np.sqrt(200*1.e-12)/2.678)
```

This means that a typical fluctuation amplitude of baryon fluctuations on these scales is 


$$\frac{\Delta\rho_{\rm bar}}{\bar{\rho}_{\rm bar}}\approx 3\frac{\Delta T}{T}\approx 10^{-5}$$

Thus, amplitude of a typical perturbation must evolve **by five orders of magnitude** from $\approx 10^{-5}$ to $\sim 1$ from $z_{\rm CMB}\approx 1100$ to $z=0$. 

### How do such density fluctuations evolve in expanding universe?

Consider a spherical perturbation of initial *physical* radius $r_i$ containing mass $M=M(<r_i)$.

<br>
<center>
<img width=500 src="img/spherical_cow.png"></img>
</center>

This simplification not only turns our problem from 3d to 1d, but also allows us to ignore the rest of the universe due to the 1st Newton's theorem (or its GR equivalent - Birkhoff's theorem). 

Assuming that as the radius $r$ of the region evolves such that the mass inside stays constant, $M=\mathrm{const}$, the density within the perturbation will evolve as 

$$\rho=\frac{3M}{4\pi r^3}$$ 

The mean density of the universe evolves
as 

$$\bar{\rho}(t)=\bar{\rho}_{\rm m0} a(t)^{-3},$$ 

where $\bar{\rho}_{\rm m0}$ is the mean matter density in the universe at the current time (redshift $z=0$) and $a(t)=(1+z)^{-1}$ is the expansion factor. 

Thus, dimensionless density contrast evolves as

$$
1+\delta(t)=\frac{\rho}{\bar{\rho}}=\frac{3M}{4\pi\bar{\rho}_{\rm m0}}\,\frac{a^3}{r^3}=R^3_{\rm L}\frac{a^3}{r^3},
$$

where $R_{\rm L}=[3M/(4\pi\bar{\rho}_{\rm m0})]^{1/3}$ is the *comoving Lagrangian radius* of the perturbation. 

The physical radius of the perturbation evolves as 

$$r(t)=R_{\rm L}\, a(t)\,[1+\delta(t)]^{-1/3}.$$


$$r(t)=R_{\rm L}\, a(t)\,[1+\delta(t)]^{-1/3}.$$

<br>
The first implication of the above equation is rather trivial: 

* When $\delta=0$ (no perturbation),  $r=R_{\rm L}a(t)$ - i.e., region simply expands with the universe at the average rate. 

### How do density perturbations evolve with time? 

Equation $r(t)=R_{\rm L}\, a(t)\,[1+\delta(t)]^{-1/3}$ is useful for understanding qualitative evolution of overdense and under-dense regions, but it contains to unknown functions, $r(t)$ and $\delta(t)$ and so by itself does not give us these functions that we would like to know. 

Thus, it needs to be solved somehow. 

Several approaches and models are frequently used. By far the most common approach, quite accurate for small perturbations ($\delta\leq 0.1$), is to differentiate this equation with respect to time once and then twice and linearize the resulting equations by neglecting terms like $1+\delta$ and terms containing powers of $\delta$ beyond linear. Such linearization procedure and derivation of relevant evolution equations is provided in the notebook.


### Initial stages of evolution: linear regime

We saw that we can write physical radius of a fluctuation at time $t$ as 

$$r(t)=R_{\rm L}\, a(t)\,(1+\delta)^{-1/3}.$$

Let's take 1st and 2nd time derivatives of $r(t)$: 

$$\dot{r}=R_{\rm L}\left[\dot{a}\,(1+\delta)^{-1/3} - \frac{a}{3}\,\dot{\delta}\,(1+\delta)^{-4/3}\right]$$

$$\ddot{r}=R_{\rm L}\left[\ddot{a}\,(1+\delta)^{-1/3} -\frac{\dot{a}}{3}\frac{\dot{\delta}}{(1+\delta)^{4/3}}- \frac{a}{3}\,\ddot{\delta}\,(1+\delta)^{-4/3}-\frac{\dot{a}}{3}\,\dot{\delta}\,(1+\delta)^{-4/3}+\frac{4\dot{a}}{9}\,\dot{\delta}^2\,(1+\delta)^{-7/3}\right]$$

For regime in which $\delta\ll 1$, so $1+\delta\approx 1$ and we can neglect quadratic terms like $\dot{\delta}^2$, we simplify the above equation to 

### Initial stages of evolution: linear regime (contd.)


$$\ddot{r}=R_{\rm L}\left[\ddot{a} -\frac{2}{3}\,\dot{a}\dot{\delta} -\frac{a}{3}\ddot{\delta}\right]$$

dividing by the expression for $r$, which in this regime is $r\approx R_{\rm L}a$: 

$$\frac{\ddot{r}}{r}=\frac{\ddot{a}}{a}-\frac{2}{3}\,\frac{\dot{a}}{a}\,\dot{\delta}-\frac{\ddot{\delta}}{3}$$

Then using $\ddot{r}=-GM/r^2$ we have: 

$$\frac{\ddot{r}}{r}=-\frac{GM}{r^3}=-\frac{4\pi G}{3}\,\bar{\rho}\,(1+\delta)=\frac{\ddot{a}}{a}-\frac{2}{3}\,\frac{\dot{a}}{a}\,\dot{\delta}-\frac{\ddot{\delta}}{3}$$

For $\delta=0$, $\dot{\delta}=0$ and $\ddot{\delta}=0$, so we have Friedman equation for expansion of the universe (region of mean density): 

$$\frac{\ddot{a}}{a} = -\frac{4\pi G}{3}\,\bar{\rho}, $$

which can be subtracted from the above equation to get the equation of overdensity evolution in the linear regime: 

$$\ddot{\delta}+2\, H(a)\dot{\delta} = 4\pi G \bar{\rho}\delta,$$

where $H(a)\equiv \dot{a}/a$ is the Hubble function. The form of this equation may remind you of equation of a commonly encountered physical system. What is it?

### Equation for the evolution of $\delta$ in the linear regime

If we use $\Omega_{\rm m}(t)=\bar{\rho}(t)/\rho_{\rm crit}(t)$, where $\rho_{\rm crit}(t)=3H(t)^2/(8\pi G)$, we can rewrite above equation in a slightly different form

$$\ddot{\delta}+2\, H(a)\dot{\delta} -\frac{3}{2}\Omega_{\rm m}(t) H^2(t)\delta = 0.$$

### Solution of the linear equation

For $\Omega_{\rm m}=1={\rm const}$, for which $a=(t/t_{\rm U})^{2/3}$, where $t_{\rm U}=2/(3H_0)$ is the current ($a=1$) age of universe. These equations combine to give $\bar{\rho}=1/(6\pi Gt^2)$, a solution of the equation is analytic (this can be checked by substitution): 

$$\delta(t) = \delta_{0+}\,\left(\frac{t}{t_U}\right)^{2/3}=\delta_{0+}\, a(t).$$

or, more generally, it can be shown (see <a href="http://adsabs.harvard.edu/abs/1977MNRAS.179..351H">Heath 1977</a> or S 4.1.6 on p. 172 in Mo et al. book): 

$$\delta(t) = \delta_{0+}D_+(t) + \delta_{0-}D_-(t),$$

where 

$$D_-(t)\propto H(t)$$

$$D_+(t)\propto H(t)\,\int\limits_0^t\frac{dt^\prime}{a^2(t^\prime)H^2(t^\prime)}\propto H(z)\int\limits_z^\infty\frac{(1+z^\prime)}{E^3(z^\prime)}\,dz^\prime$$

For the $\Omega_{\rm m}=1,\ \Omega_\Lambda=0$ cosmology 
$$\delta_+\propto t^{2/3}\propto a,$$
and so in the linear regime overdensity increases proportionally to expansion factor.

At the same time, $\delta_-\propto H(t)\propto t^{-1}$. Thus, $\delta_+$ grows with time, while $\delta_-$ decreases with time. Hence, $D_+$ is called *the growth function.* 

By late time epochs relevant for structure formation the second term in the above sum thus is small compared to the first. The presence of the two terms tells us that only a fraction $\delta_{0+}$ of the originally existing fluctuation $\delta_0$ survives, while another fraction decays due to expansion of the universe. 


### Growth of linear perturbations in other cosmologies 

In other cosmologies, solution has the same form

$$\delta(t) = \delta_{0+}D_+(t) + \delta_{0-}D_-(t),$$

where 

$$D_-(t)\propto H(t);\ \ \ \ D_+(t)\propto H(t)\,\int\limits_0^t\frac{dt^\prime}{a^2(t^\prime)H^2(t^\prime)}\propto H(z)\int\limits_z^\infty\frac{(1+z^\prime)}{E^3(z^\prime)}\,dz^\prime$$

but the integral in expression for $D_+$ is generally not analytic and needs to be evaluated numerically. It depends on $H(t)$ and dimensionless Hubble function

$$E(z)\equiv \frac{H(z)}{H_0}=\sqrt{\Omega_{\rm m,0}\,(1+z)^3+\Omega_k\,(1+z)^2+\Omega_{\Lambda,0}},$$
where $\Omega_k=1-\Omega_{\rm m,0}-\Omega_\Lambda$ and so the growth function depends on cosmological parameters through $H(z)$ and $E(z)$.  

The exception are "open geometry" cosmologies, i.e. $\Omega_{\rm m,0}<1$ and $\Omega_\Lambda=0$, for which the growth factor can be written in closed form (see eq. 4.74 in the Mo et al. book). 

It is customary to normalize $D_+(t)$ to unity at the present epoch ($z=0$), so that linear evolution of the growing part of the perturbation can be written as

$$\delta \equiv \delta_{0+}\,D_+(t,\Omega_{\rm m},\Omega_\Lambda).$$

Subscript in $\delta_{0+}$ emphasizes that only the growting part of the original primordial amplitude of the fluctuations survives to present epoch, while part of it decayed due to the decaying mode of the solution $\propto D_-(t)$. 

### How do density perturbations evolve with time? 

The solution of linearized equation for $\delta(t)$ that is growing with time can be expressed by 

$$\delta(t) \propto D_+(t), $$

where $D_+(t)$ is the *linear growth factor* function that depends on the cosmological parameters describing background cosmology. 

The fastest growth of $\delta$ occurs for $\Omega_{\rm m0}=1$ model: 

$$\delta(t) \propto a(t)$$

### Evolution of perturbation amplitude between $z_{\rm CMB}$ and $z=0$ is thus

$$\frac{\delta(z=0)}{\delta(z_{\rm ls})}=\frac{1}{a_{\rm ls}}=(1+z_{\rm ls})\sim 10^3.$$


Thus, $\delta(z_{\rm CMB})\sim 10^{-5}$ perturbations in baryon density revealed by the CMB temperature fluctuations would grow only to $\delta(z=0)\sim 10^{-2}$ fluctuations during the age of the universe. 

However, large-scale distribution of galaxies reveals order unity amplitude of density fluctuations on relevant scales.

### What's going on?

How can we reconcile the level of baryon density perturbations revealed by the CMB temperature fluctuations and amplitude of density fluctuations in galaxy distribution?  

In principle, if we did not know anything we could suspect break down of linear model, or that galaxy distribution is highly biased (i.e. have much larger fluctuation amplitude than underlying matter).

However, it turns out that both these possibilities are not tenable: linear model is still quite good at $\delta\approx 10^{-2}$ and breaks down strongly only at $\delta\sim 1$, while although galaxy distribution is indeed biased relative to distribution of matter, the difference (bias) is of order unity, not two orders of magnitude. 

### Dark matter to the rescue 

Postulate existence of gravitationally dominant matter component that does not interact electromagnetically (and therefore does not interact with baryons and does not emit EM radiation - i.e. "dark"), which has $\delta(z_{\rm CMB})\sim 10^{-3}$.

### Evolution of overdense and underdense regions in the $\Lambda$CDM cosmology

We saw that we can write physical radius of a fluctuation at time $t$ as 

$$r(t)=R_{\rm L}\, a(t)\,[1+\delta(t)]^{-1/3}.$$


If at some redshift $z_i$ (or expansion factor $a_i=(1+z_i)^{-1}$) perturbation has amplitude $\delta_i$, then equation implies: 

$$r_i=R_{\rm L}\, a_i\,(1+\delta_i)^{-1/3}.$$

This means: 

* When fluctuation is negative, $-1<\delta_i<0$, $r_i>R_{\rm L}\, a(t_i)$ - i.e., at $t<t_i$ fluctuation was expanding faster than region of average density of the same mass. 

* When fluctuation is positive, $\delta_i>0$, $r_i<R_{\rm L}a(t_i)$ - i.e., at $t<t_i$ fluctuation was expanding slower than the universe.

### What are the qualitative consequences of such evolution of $\delta>0$ and $\delta<0$ regions? 

Recall that initial *Gaussian pdf of $\delta$ is symmetric around its mean $\langle\delta\rangle=0$.* 

This implies an important "topological" property of the initial Gaussian $\delta$ field...

The fraction of volume in $\delta>0$ and $\delta<0$ regions is the same. 

### Consequence 1: 

Fraction of volume occupied by underdense regions will grow, while fraction of volume of overdense regions will decrease. 

Consequently, gravity and *gravitational instability will turn a Gaussian $\delta$ field into a non-Gaussian field.*

Neighboring under-dense regions - i.e. underdense regions separated by an overdense one can "run into" each other. The interface between them is a "flattened" sheet-like structure.

Note that temperature fluctuations exist on a wide range of scales. 

<center>

How can they be characterized?

### Observed temperature fluctuations of the CMB are consistent with being an isotropic  Gaussian random field

Given the linear relation of $\Delta T/\bar{T}$ and $\delta_{\rm b}=\Delta\rho_{\rm b}/\bar{\rho}_{\rm b}$, this implies that $\delta_{\rm b}$ field is isotropic and Gaussian. 

What do isotropy and Gaussianity of perturbations mean operationally?

 First of all, Gaussianity implies that probability distribution function (pdf) of matter *overdensity* $\delta=\rho/\bar{\rho}-1$ is Gaussian with the mean $\langle\delta\rangle=0$ (by definition of $\delta$): 

$$
p(\delta)=\frac{1}{\sqrt{2\pi}\sigma}\exp\left[-\frac{\delta^2}{2\sigma^2}\right].
$$

where $\sigma^2$ is the variance of the $\delta$ field. 

Variance of the Gaussian pdf in general depends on the scale in the sense that if we filter out perturbations of size smaller than some scale $R$ using a filter and measure rms of the filtered $\delta_R$ field:

  
$$\delta_{R}({\bf x})\equiv \int \delta({\bf x}-{\bf r})W({\bf r}, R)d^3\mathbf{r},$$ 


its *rms* will depend on $R$:

$$\sigma_R =\sigma(R)$$



An important property of an isotropic Gaussian field is that $\sigma(R)$, where $R$ is scalar, fully specifies the field statistically. 

Note that Gaussian density perturbation field smoothed using a linear filter is also Gaussian with the probability distribution function
(PDF) given by 

$$
p(\delta_{R})=\frac{1}{\sqrt{2\pi}\sigma(R)}\exp\left[-\frac{\delta^2_{R}}{2\sigma^2(R)}\right].
$$

The scale $R$ is often replaced by the corresponding mass scale, $M$ - the total mass of matter mass "encompassed" by the filter. 

### Tophat filter

Conceptually, the meaning of $M$ and interchangeability of $M$ and $R$ are simplest for the real-space spherical tophat window: 

$$W_{\rm TH}(\mathbf{x}, R)=\frac{3}{4\pi R^3}\,
\begin{cases} 1, \vert\mathbf{x}\vert\leq R,\\ 
0, \vert\mathbf{x}\vert > R
\end{cases}$$

for which 

$$M=\frac{4\pi}{3}\,\bar{\rho}_{\rm m}(z)R^3,$$  

where $\bar{\rho}_{\rm m}(z)$ is the mean density of matter in the universe at redshift $z$. $M$ and $R$ are thus interchangeable and can be used to specify filter using a size or mass scale.

The Fourier counterpart of the tophat filter is  

$$\tilde{W}_{\rm TH}(x)=\frac{3(\sin x-x \cos x)}{x^3}, \ \ \ {\rm where}\ \ x=kR,$$

### Variance of the field and its relation to power spectrum $P(k)$

Formal expression for $\sigma(R)$ of the filtered $\delta_R$ is: 

$$
\sigma^2(R) = \langle\delta_{R}^2\rangle=\frac{1}{(2\pi)^3}\int
P(k)\vert \tilde{W}({\bf k},R)\vert^2d^3\mathbf{k},
$$ 

where $P(k)$ is the *power spectrum*, defined as the volume average of the square of Fourier harmonic amplitudes: $P(k)=\langle\vert\delta_{\mathbf{k}}\vert^2\rangle$; $d^3\mathbf{k}=4\pi k^2dk$ is the volume element in Fourier space, and $\tilde{W}({\bf k},R)$ is the Fourier transform of the window
(filter) function $W({\bf r},R)$. 

### Cross-scale moments of $\delta$

Sometimes it is also useful to define moments of the distribution at two different scales, $r$ and $R$: 

$$\langle\delta_r\delta_R\rangle = \frac{1}{(2\pi)^3}\int P(k)\vert \tilde{W}(k,r)\tilde{W}(k,R)\vert\,d^3\mathbf{k}.$$

For example, we will later use such moment to compute mean overdensity, $\delta(<r)$, around points conditioned that $\delta(<R)$ is equal to some value $\delta_{R\ast}$ at some other scale $R$.

### Power spectrum of fluctuation amplitudes

The power spectrum can also be defined via total variance of the unsmoothed density field $\langle\delta^2\rangle$:

$$
\langle\delta^2\rangle=\frac{1}{(2\pi)^3}\int P(k)\,d^3\mathbf{k}=\frac{1}{2\pi^2}\int P(k)k^2 dk, 
$$

Note that $\delta^2$ is dimensionless and the above expression shows that $P(k)$ has dimensions of $k^{-3}$ - i.e., dimensions of volume and is not the most intuitive quantity physically. 

### Dimensionless power spectrum 

However, we can define dimensionless power spectrum, $\Delta^2(k)$, as

$$
\langle\delta^2\rangle=\frac{1}{2\pi^2}\int P(k)k^2 dk 
= \frac{1}{2\pi^2}\int P(k)k^3 d\ln k = \int\Delta^2(k)d\ln k, 
$$

or 
$$\Delta^2(k)=\frac{4\pi k^3}{(2\pi)^3}P(k)=\frac{1}{2\pi^2}k^3 P(k).$$

which does have an intuitive meaning: $\sqrt{\Delta^2(k)}$ is a typical amplitude of fluctuations, $\delta_k$, over a logarithmic interval around given $k$. 

### Spectrum of the CMB temperature measured by the Planck satellite

A similar dimensionless power spectrum can be defined for the CMB temperature fluctuations in the angular sky coordinates: 

$$\frac{\Delta_{\rm T}^2}{\langle T\rangle^2}\,\equiv \frac{l(l+1)}{2\pi}\, C_l $$

where $\Delta_{\rm T}^2$ is in $(\mu\rm K)^2$ and $\langle T\rangle=2.768\ \rm K$ and $l$ is multipole moment $l\approx 180^\circ/\theta$ corresponding to the angular scale $\theta$.

Note that $\Delta(k)=\sqrt{\Delta^2(k)}$ is characteristic amplitude $\delta$ of density fluctuations. We can see that at $k\sim 0.1-1h\,\rm Mpc^{-1}$, $\Delta\sim 1-3\times 10^{-3}$ - sufficient to grow to $\delta\sim 1$ fluctuations at $z=0$.

### Realization of the Gaussian $\delta$ field with the specified power spectrum $P(k)$

The code below generates a random Gaussian realization of the $\delta$ field on a cubic grid of specified size $N$ with the power spectrum $P(k)$ computed for a given cosmological model using the <a href="https://bdiemer.bitbucket.io/colossus/cosmology_cosmology.html#cosmology.cosmology.Cosmology.matterPowerSpectrum">Colossus' power spectrum routine</a>. 

A slide through the $\delta$ field is then plotted. 

```python
import numpy as np
from numpy.fft import fftn, ifftn, fftfreq
from numpy.random import normal
from matplotlib import pylab as plt

from colossus.cosmology import cosmology

# define a vector of cosmological parameters:    
my_cosmo = {'flat': True, 'H0': 70.0, 'Om0': 0.3, 'Ob0': 0.045714, 'sigma8': 0.82, 'ns': 0.96}
# set my_cosmo to be the current cosmology
cosmo = cosmology.setCosmology('my_cosmo', my_cosmo)
cosmo.interpolation = False

# box size in h^{-1} Mpc
Lbox = 100.0
# grid size (change N to 128 if 256 is taking too long)
N=256; N3 = N*N*N
dx = Lbox/N; dxi = 1.0/dx

# set up grid coordinates
x1d = np.linspace(0, N-1, N)
x, y, z = np.meshgrid(x1d, x1d, x1d)

# set random seed; remove if you want new realization every time
np.random.seed(141592653)
# generate 3D array of Gaussian numbers with zero mean and unit dispersion
delta = np.reshape(normal(0.0, 1.0, N3),(N,N,N))

# Fourier transform delta field
delta = fftn(delta)
# get vector of wavenumbers corresponding to modes, fftfreq takes care of mode "packing" order
k1d = fftfreq(N) * 2.0 * np.pi
# grid of k in Fourier space
kx, ky, kz = np.meshgrid(k1d, k1d, k1d)
k = dxi * np.sqrt(kx**2 + ky**2 + kz**2)
k[0,0,0] = np.pi * 2.0 / Lbox

# set up power spectrum for the input model at redshift zini
zini = 30.0
# growth factor at 30. and its square
gzi = cosmo.growthFactor(30.0); gzi2 = gzi * gzi
# normalize the power spectrum to z=30
Pk  = cosmo.matterPowerSpectrum(k) * gzi2 

# normalize Fourier components using sqrt(P(k))
delta *= np.sqrt(Pk)
# this is "DC" or "fundamental mode" corresponding to constant density in the box. 
# it can be set to any constant
delta[0,0,0] = 0.0
```

```python
Nslice = int(N/2)

plt.figure(figsize=(3,3))
plt.xlim(0., Lbox); plt.xlim(0., Lbox)
plt.xlabel(r'$x\ (h^{-1}\rm\ Mpc)$')
plt.ylabel(r'$y\ (h^{-1}\ \rm Mpc)$')
plt.imshow(delta[:,:,Nslice], cmap='winter', vmin=dmin, vmax=dmax, extent=(0., Lbox, 0., Lbox))
#plt.savefig('img/gfield_unsmoothed.png', bbox_inches='tight')

plt.show()
```

The code below computes power spectrum of the $\delta$ field on a grid and compares it with the input model spectrum in a plot as a consistency check. 

```python
# compute and check P(k)
lk = np.log10(k)
Nb = 20 # number of k bins to use
# P(k) itself - is just a weighted histogram of squares of delta_k amplitudes
Pkj,  lkj = np.histogram(lk, bins=Nb, weights=np.square(np.absolute(delta)), density=False) 
# <delta^4> for variance of the P(k)
Pkj2, lkj = np.histogram(lk, bins=Nb, weights=np.square(np.square(np.absolute(delta))), density=False) 
lkaj, lkj = np.histogram(lk, bins=Nb, weights=lk, density=False) 
nj , lkj  = np.histogram(lk, bins=Nb, density=False) 
# P(k) and its rms error due to finite number of modes
Pkj = Pkj / nj / gzi2 / N3
Pkj2 = Pkj2 / (gzi2*gzi2) / N3 / N3
sPkj = np.sqrt(Pkj2/(nj*(nj-1)) - Pkj**2/(nj-1))
lkaj = lkaj / nj 
kaj = 10.**lkaj

# transform the field back into real space
delta = np.real(ifftn(delta))

# input model spectrum for comparison
lka = np.arange(-4.,3.,0.02); ka = 10.0**lka
Pka = cosmo.matterPowerSpectrum(ka)

#
# plot input model P(k) and P(k) for the realization
#
Pkjp = np.clip(Pkj,1.e-10,1.e100)
plt.figure(figsize=(3,3))
plt.xscale('log'); plt.yscale('log')
plt.ylim(1.e-2, 1.e5); plt.xlim(1.e-4, 1.e2)
plt.errorbar(kaj, Pkjp, yerr=2.*sPkj, fmt='o', markersize=1.)
plt.scatter(kaj, np.clip(Pkj,1.e-10, 1.e100), s=10., label=r'$\mathrm{realization}$')
plt.plot(ka, Pka, c='slateblue', linewidth=2., label=r'$\mathrm{input\ model}$')
plt.legend(frameon=False, loc='lower left', fontsize=10, scatterpoints=1)
plt.xlabel(r'$k\ (h\rm\ Mpc^{-1})$')
plt.ylabel(r'$P(k)\ (h^{-3}\ \rm Mpc^3)$')
#plt.grid()
plt.show()
```

The code below uses the $\delta$ field generated on a grid to smooth it with the tophat window filter of three different radii and plot the resulting smoothed fields $\delta_R$. 

```python
from scipy import signal

def Wfk(kR, filter = 'tophat'):
    """
    Top-hat and Gaussian window functions in the Fourier space
    """
    if filter == 'tophat':
        return 3.*(np.sin(kR)-kR*np.cos(kR))/(kR)**3
    elif filter == 'gaussian':
        return np.exp(-0.5*kR**2)
    else:
        print("error in Wfk: routine accepts only tophat or gaussian filter")
        return

#
# plot a slice of the density field of Gaussian realization to check things visually
#
import matplotlib.ticker as mtick

lloc = 'upper right'; lfs = 9; lcol = 'white'; ccol = 'yellow'
fig, ax = plt.subplots(2,2,figsize=(4,4))

for ix in [0,1]:
    for iy in [0,1]:
        ax[ix][iy].set_xlim(0., Lbox); ax[ix][iy].set_ylim(0., Lbox)

Nslice = int(N/2)

dmin = 0.75*np.min(delta[:,:,Nslice]); dmax = 0.5*np.max(delta[:,:,Nslice])
x1d = Lbox*np.linspace(0, N-1, N)/(N-1)
x, y = np.meshgrid(x1d, x1d)

ax[0][0].imshow(delta[:,:,Nslice], cmap='winter', vmin=dmin, vmax=dmax, extent=(0., Lbox, Lbox, 0.))
ax[0][0].plot(0, 0, "-", c='black', label=r'$\mathrm{no\ smoothing}$' )
#ax[0][0].contour(x, y, delta[:,:,N/2]/gzi.T, extent=(0., Lbox, 0., Lbox), linewidths=(0.5), colors=ccol, levels = [1.69])

#ax[0][0].colorbar(fraction=0.04, pad=0.04)
lgnd = ax[0][0].legend(frameon=False, loc=lloc, fontsize=lfs)
for text in lgnd.get_texts():
    text.set_color(lcol);
for item in lgnd.legendHandles:
    item.set_visible(False)
 
# smooth the field with a given filter size and plot
Rfa = [0.5, 1.5, 5.0] #/h Mpc
ix = [0, 1, 1]; iy = [1, 0, 1]
# smooth by a real-space top-hat filter
filter = 'tophat'

for ir, Rf in enumerate(Rfa):
    kR = k*Rf; 
    deltas = fftn(delta) * Wfk(kR, filter)
    deltas = np.real(ifftn(deltas))
    ax[ix[ir]][iy[ir]].imshow(deltas[:,:,Nslice], vmin=dmin, vmax=dmax, 
                              cmap='winter',  extent=(0., Lbox, Lbox, 0.) )
    delta_s = deltas[:,:,Nslice]/gzi
    ax[ix[ir]][iy[ir]].contour(x, y, delta_s,  linewidths=(0.5), colors=ccol, 
                               levels = [1.69])
    ax[ix[ir]][iy[ir]].plot(0, 0, "-", c='black', label=r'$R_{\rm f}=%.1fh^{-1}\ \mathrm{Mpc}$'%Rf )
    lgnd = ax[ix[ir]][iy[ir]].legend(frameon=False, loc=lloc, fontsize=lfs)
    for text in lgnd.get_texts():
        text.set_color(lcol);
    for item in lgnd.legendHandles:
        item.set_visible(False)

plt.setp(ax[0][0].get_xticklabels(), visible=False)
plt.setp(ax[0][1].get_xticklabels(), visible=False)
plt.setp(ax[0][1].get_yticklabels(), visible=False)
plt.setp(ax[1][1].get_yticklabels(), visible=False)
ax[0][0].set_ylabel(r'$y\ (h^{-1}\ \mathrm{Mpc})$')
ax[1][0].set_ylabel(r'$y\ (h^{-1}\ \mathrm{Mpc})$')
ax[1][0].set_xlabel(r'$x\ (h^{-1}\ \mathrm{Mpc})$')
ax[1][1].set_xlabel(r'$x\ (h^{-1}\ \mathrm{Mpc})$')

plt.tight_layout()

#plt.savefig('img/gfield_smooth.png', bbox_inches='tight')
plt.show()
```
